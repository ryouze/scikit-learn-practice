# 1. Basic Model
This notebook introduces the basic workflow of building a regression model with scikit-learn. We use the Boston housing dataset to predict house prices from various features. You'll see how to load data, select and fit a model, make predictions, and visualize results.

```python
# Load the Boston housing dataset (features and target values)
from sklearn.datasets import load_boston
```

```python
# X contains the input features, y contains the target (house prices)
X, y = load_boston(return_X_y=True)
```

```python
# Import regression models: KNeighborsRegressor and LinearRegression
# from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Choose a regression model. You can easily swap models thanks to scikit-learn's unified API.
mod = KNeighborsRegressor()

# The beauty of sci-kit learn is that the API is the same for all models
# mod = LinearRegression()
```

```python
# This won't work because the model is not fitted yet
# mod.predict(X)

mod.fit(X, y)
```

```python
# Predict house prices for the training data
pred = mod.predict(X)
```

## Visualizing Predictions vs. True Values
Let's compare the predicted prices to the true prices. If the model is accurate, the points will be close to the diagonal line.

**Note:** KNeighborsRegressor predicts by averaging the k (default: 5) nearest neighbors in the training set.

```python
# Import matplotlib for plotting
import matplotlib.pylab as plt
```

```python
plt.scatter(pred, y)
```

## Why Feature Scaling Matters
KNeighborsRegressor uses distance between feature vectors to find neighbors. If features are on different scales (like 'square feet' in thousands and 'proximity to school' as 0/1), unscaled data can bias the results.

To avoid this, it's important to preprocess features so each one contributes fairly. In the next notebook, we'll see how to scale features before fitting the model.
